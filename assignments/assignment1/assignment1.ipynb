### Task 1: A list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

#### a) Filter invalid transactions

```python
valid_transactions = [
    t for t in transactions if t["price"] > 0 and t["quantity"] >= 0
]
```

#### b) Compute total

```python
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]
```

#### c) Group by region and sum revenue

```python
revenue_by_region = {}

for t in valid_transactions:
    region = t["region"]
    revenue_by_region[region] = revenue_by_region.get(region, 0) + t["total"]
```

#### d) Find region with highest revenue

```python
max_region = max(revenue_by_region, key=revenue_by_region.get)
max_value = revenue_by_region[max_region]
```

```python
print(f"Highest revenue region: {max_region} with total {max_value}")
```

## Task 2: Create a function

```python
def summarise_sales(sales):
    total = sum(sales)
    average = round(total / len(sales), 2)
    
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) + 1
    
    above_average_count = sum(1 for s in sales if s > average)
    
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }


sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

result = summarise_sales(sales)
print(result)
```

## Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

#### a) Passing students

```python
passing = []

for s in students:
    if s["grade"] >= 50:
        passing.append(s["name"])

print("Passing:", passing)
```

#### b) Status messages

```python
for s in students:
    name = s["name"]
    grade = s["grade"]
    
    if grade >= 80:
        print(f"{name}: Excellent")
    elif grade >= 50:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")
```

#### c) Count low attendance

```python
low_attendance_count = 0

for s in students:
    if s["attendance"] < 0.80:
        low_attendance_count += 1

print("Low attendance count:", low_attendance_count)
```

#### d) Highest grade using loop

```python
top_student = students[0]

for s in students:
    if s["grade"] > top_student["grade"]:
        top_student = s

print(f"Top student (loop): {top_student['name']} with {top_student['grade']}")
```

#### e) Highest grade using max()

```python
top_student_max = max(students, key=lambda s: s["grade"])

print(f"Top student (max): {top_student_max['name']} with {top_student_max['grade']}")
```

### Task 4: Class and methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}

    # a) Method that adds new or updates asset
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }

    # b) Method total value
    def total_value(self):
        total = 0
        for data in self.holdings.values():
            total += data["shares"] * data["price_per_share"]
        return total

    # c) Method most valuable
    def most_valuable(self):
        max_ticker = None
        max_value = 0
        for ticker, data in self.holdings.items():
            value = data["shares"] * data["price_per_share"]
            if value > max_value:
                max_value = value
                max_ticker = ticker
        return max_ticker

    # d) A __str__ method
    def __str__(self):
        return f"Portfolio of {self.owner}: {len(self.holdings)} assets, Total value = {self.total_value()}"
```

```python
p = Portfolio("Anh")
p.add_asset("AAPL", 10, 150)
p.add_asset("GOOGL", 5, 2800)
p.add_asset("TSLA", 8, 700)

p.add_asset("AAPL", 5, 160)

print(p)
print(p.total_value())
print(p.most_valuable())
```

### Task 5: Importing a module

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Parse dates into datetime.date objects
dates = []

for date_str, product in orders:
    date_obj = datetime.strptime(date_str, "%Y-%m-%d").date()
    dates.append(date_obj)

# Find earliest and latest
earliest = min(dates)
latest = max(dates)

# Compute difference in days
difference = (latest - earliest).days

print(f"Days between earliest and latest order: {difference}")
```

### Task 6: Array Operations and Broadcasting

```python
import numpy as np

# Data
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])
```

#### a) Total annual revenue (row sums)

```python
annual_totals = np.sum(revenue, axis=1)
print("Annual revenue per year:", annual_totals)
```

#### b) Mean quarterly revenue (column means)

```python
quarterly_means = np.mean(revenue, axis=0)
print("Mean revenue per quarter:", quarterly_means)
```

#### c) Broadcasting growth adjustment for 2024

```python
growth = np.array([1.05, 1.03, 1.08, 1.02])

adjusted_2024 = revenue[2] * growth
print("Adjusted 2024 revenue:", adjusted_2024)
```

#### d) Values greater than 250

```python
high_values = revenue[revenue > 250]
print("Values > 250:", high_values)
```

### Task 7: Pandas

```python
import pandas as pd

# Create DataFrame
data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

df = pd.DataFrame(data)
```

#### a) Add new column

```python
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)
print("a) New column:\n", df)
```

#### b) Sort and get top 3

```python
top3 = df.sort_values(by="performance_score", ascending=False).head(3)
print("\nb) Top 3 employees:\n", top3)
```

#### c) Filter with .loc

```python
finance_high_perf = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
]
print("\nc) Finance + high performance:\n", finance_high_perf)
```

#### d) Rename column

```python
df = df.rename(columns={"years_exp": "experience_years"})
print("\nd) Renamed column:\n", df.columns)
```

#### e) Select with .iloc

```python
subset = df.iloc[:4][["salary", "performance_score"]]
print("\ne) First 4 employees (selected columns):\n", subset)
```

### Task 8: More Pandas

```python
import numpy as np
import pandas as pd

# Generate data
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

#### a) Create DataFrame and set index

```python
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df = df.set_index("date")
print("a) DataFrame:\n", df.head())
```

#### b) Create new columns

```python
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round().astype(int)
df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

print("\nb) Updated DataFrame:\n", df.head())
```
